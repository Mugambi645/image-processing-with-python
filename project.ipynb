Transforming images to grayscale

```python
#Import the modules from skimage - A python module for image processing
from skimage import color, data, io
#load the image
rocket = data.rocket()
#Grayscale conversion is a common preprocessing step, as it simplifies image data while preserving essential information.
#Scikit-image provides the color.rgb2gray() function to perform this conversion.
gray_scaled_rocket = color.rgb2gray(rocket)
io.imshow(rocket)
```

```python
io.imshow(gray_scaled_rocket)
```

Aliasing, rescaling and rotating images

```python
#Import rotate and rescale functions
from skimage.transform import rotate, rescale
```

```python
#Import the image
cat_image = plt.imread("Image Processing with Python course exercise dataset/chapter 2/image_cat.jpg")
# Rotate the image 90 degrees clockwise
rotated_cat_image = rotate(cat_image, -90)
#rescale with anti-aliasing
rescaled_with_aa = rescale(rotated_cat_image, 1/4, anti_aliasing=True)
rescaled_without_aa = rescale(rotated_cat_image, 1/4, anti_aliasing=False)
```

```python
io.imshow(rescaled_with_aa)
```

```python
io.imshow(rescaled_without_aa)
```

```python
io.imshow(rotated_cat_image)
```

```python
io.imshow(cat_image)
```

 Image Histogram Equalization

Histogram equalization improves image contrast by redistributing pixel intensity values. Scikit-image provides the exposure module to perform histogram equalization and related operations.

```python
from skimage import exposure
gray_image = color.rgb2gray(cat_image)
equalized_image = exposure.equalize_hist(gray_image)
io.imshow(equalized_image)
io.show()
```

- Colors in Numpy, colors in histograms
  
now let's represent each channel(R,G,B) as an array and see the its image

```python
from skimage.data import astronaut
as_image = astronaut()
red_as = as_image[:, :, 0]
green_as = as_image[:, :, 1]
blue_as = as_image[:, :, 2]
fig = plt.figure(figsize=(15, 5))
for i, j, k in zip([as_image, red_as, green_as, blue_as], range(4), ["original", "red", "green", "blue"]):
    ax = fig.add_subplot(1, 4, j+ 1)
    plt.imshow(i)
    plt.title(k)
plt.show()
```

```python
fig=plt.figure(figsize=(15,3))
for i,j,k in zip([red_as, green_as, blue_as], range(3), ['red', 'green', 'blue']):
    ax=fig.add_subplot(1,3,j+1)
    plt.hist(i.ravel(), color=k, bins=256)
    plt.title(k)
plt.show()
```

Threshholding

partitioning an image into a foreground and background by making it black and white, and this is the simplest method of image segmentation
skimage has alot of filters to apply thresholding

```python
from skimage.data import coffee
from skimage.filters import try_all_threshold, threshold_otsu, threshold_local, sobel, gaussian
```

```python
io.imshow(coffee())
```

```python
# Optimal thresholding
coffee_image = coffee()
#convert to grayscale
gray_coffee_image = color.rgb2gray(coffee_image)
threshold = threshold_otsu(gray_coffee_image)
thresh_img = gray_coffee_image > threshold
plt.title("Effect of optimal thresholding")
io.imshow(thresh_img)
```

```python
#local thresholding
from skimage.data import page
page_image = page()
local_thresh = threshold_local(page_image, 51)
local_thresh_img = page_image > local_thresh
plt.title("Effect of local thresholding")
io.imshow(local_thresh_img)
```
